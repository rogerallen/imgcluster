```python
image_directory = "./images"

clusterer = ImageClusterer(image_directory)
clusterer.load_features("my_image_features.pkl")

clusterer2 = ArtisticImageClusterer(image_directory, model_name="ViT-B/32")
clusterer2.load_features("artistic_vitb32_image_features.pkl")
```
